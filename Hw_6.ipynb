```python
#детальна інформація про найменш і найбільш прибуткових фільмах 
max_profit = df['Profitability'].max()
min_profit = df['Profitability'].min()
print(max_profit)
print(min_profit)
```

```python
#За допомогою loc[] ми отримуємо доступ до групи рядків і стовпців по Profitability, а isin() вибрала лише ті рядки які мають min_profit
df.loc[df['Profitability']. isin([min_profit])] 
```

```python
df.loc[df['Profitability']. isin([max_profit])] 
```

```python
#Значення "Жанр"  інколи здається непослідовним; знайти ці невідповідності і виправити їх  
genre_df = df.groupby('Genre').count()
genre_df
```

- Як бачимо, що в нас є три групи Comdy, Comedy і comedy і це явно один жанр, в першій просто пропущено одну літеру, а в третя написана з маленької букви. Тому замінемо Comdy і  comedy на Comedy
- Також міняємо Romence, romance на Romance

```python
#використаємо функцію replace()
df = df.replace({'Genre':{'Comdy':'Comedy','comedy':'Comedy', 'Romence':'Romance', 'romance':'Romance'}})
```

```python
genre_df = df.groupby('Genre').count()
genre_df
```

```python
#Зберегти (в новий файл csv) 10 найкращих комедій по кількості глядачів; показати лише назву фільма, рік і студію
best_comedy = df.sort_values('Audience score %', ascending=False)
best_comedy
```
