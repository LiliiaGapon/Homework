1. Знайдіть рекорди для кожної статі та кожного підрозділу в кожному з 3-х -
жим лежачи(best3bench_kg), присідання(best3squat_kg) та станова

2. Загальна кількість перемог кожного, приймаючи за перемогу 1-е місце. Збережіть результати у вторинному DataFrame.

```python
winners = data[data['place'] == '1'].groupby('name').agg(winners_number=('place', 'count'))
winners
```

```python
winners.to_csv('all_winners.csv')
```

3. Використовуйте DataFrame з вправи 2, щоб показати для кожного наявність підрозділу та статті, наявного з серед учасників, які будь-коли брали участь у цьому дивізіоні.

```python
data = data.merge(winners, how='left', on='name') #До data доєднуємо датафрейм winners і заповнюємо по кожному його загальну кількість перемог
data
```

```python
division_best = data.groupby(['division', 'weight_class_kg', 'sex']).winners_number.max()
print(division_best)
```
